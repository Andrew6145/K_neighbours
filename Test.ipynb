```python
#запускать эту ячейку, только если хотите маштабировать значения от 0 до 1
import math
#df.index.name="Index column"
def scaling(massive):
    max=-1
    for element in massive:
        if element>max:
            max=element
    #print(max)
    min=max
    for element in massive:
        if element<min:
            min=element
    #print(min)
    range=max-min
    #print(range)
    i=0
    for element in massive:
        element_in_range=massive[i]-min
        #print(element_in_range/range)
        massive[i]=element_in_range/range
        i+=1
    print(massive,"-array")
    return massive
scaling(rok)
scaling(pop)
scaling(dzas)
df
```

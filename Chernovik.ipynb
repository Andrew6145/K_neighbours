```python
import pandas as pd 
import random
df= pd.read_csv("data.csv.csv")
df.head()



import numpy as np
new_rok=df['acousticness'].to_numpy()
new_dzas=df["danceability"].to_numpy()
pop=df["duration_ms"].to_numpy()
new_pop=np.array([])
okay=0
error=0
for element in pop:
    element=str(element)
    element=element[:-2]
    try:
        element=int(float(element))
        new_pop=np.append(new_pop,element)
        okay+=1
    except ValueError:
        error+=1
#print("Все норм в ",okay," случаях")
#print("Ошибка в ",error," случаях")
#print(new_pop[0:5])
#print(new_rok[0:5])
#print(new_dzas[0:5])




import math
df.index.name="Index column"
rasstoyaniye=np.array([])
def scaling(massive):
    max=0
    for element in massive:
        if element>max:
            max=element
    min=max
    for element in massive:
        if element<min:
            min=element
    range=max-min
    i=0
    for element in massive:
        massive[i]=massive[i]-min
        massive[i]=massive[i]/range
        i+=1

scaling(new_rok)
scaling(new_pop)
scaling(new_dzas)

our_user_data=np.array([])
i = df.index.get_loc(3)
our_user_data=df.iloc[i]
our_user_data=our_user_data[1:4]
our_user_data=our_user_data.to_numpy()


def rasstoyaniye(masive1,massive2,massive3):
    rasstoyaniye_massive=np.array([]) 
    for element in new_pop:
        try:
            index_elementa=np.where(massive3 == element)
            calculations=math.sqrt((our_user_data[0]-masive1[index_elementa])**2+(our_user_data[1]-massive2[index_elementa])**2+(our_user_data[2]-massive3[index_elementa])**2)
            rasstoyaniye_massive= np.append(rasstoyaniye_massive,calculations)
        except TypeError:
            continue

    k_nearest_neighbours=np.array([])
    k_nearest_neighbours=np.array([random.randint(100000000000,100000000000000000000) for i in range(4)])
    #print(k_nearest_neighbours)
    k_nearest_neighbours_indexes=np.array([random.randint(100000000000,100000000000000000000) for i in range(4)])
    for element in rasstoyaniye_massive:
        #нужно получить массив индексов
        massive_indexov_posle_sortirovki=np.argsort(k_nearest_neighbours)
        k_nearest_neighbours=np.sort(k_nearest_neighbours)
        #теперь нужно поставить индексы на правильные места, для соответсвия значениям
        counter_of_indexes=0
        for correct_index in massive_indexov_posle_sortirovki:
            index_of_element_on_this_place=massive_indexov_posle_sortirovki[counter_of_indexes]
            k_nearest_neighbours_indexes[counter_of_indexes]=k_nearest_neighbours_indexes[index_of_element_on_this_place]
            counter_of_indexes+=1
        max_for_k_nearest_neighbours=k_nearest_neighbours[-1]
        if element<max_for_k_nearest_neighbours and element!=k_nearest_neighbours[0] and element!=k_nearest_neighbours[1] and element!=k_nearest_neighbours[2] and element!=k_nearest_neighbours[3]:
            for value in k_nearest_neighbours:
                index_elementa=np.where(k_nearest_neighbours == value)
                print(index_elementa,"Это индекс элемента")
                if element<value:
                    k_nearest_neighbours[index_elementa]=element
                    indexik_elementa=np.where(rasstoyaniye_massive == element)
                    k_nearest_neighbours_indexes[index_elementa]=indexik_elementa
                    break
    #print(rasstoyaniye_massive)
    print(k_nearest_neighbours,"-это расстояния между 4 ближайшеми соседями")
    print(k_nearest_neighbours_indexes,"-это индексы 4 ближайших к пользователю соседей в таблице")
rasstoyaniye(new_rok,new_dzas, new_pop)

```
